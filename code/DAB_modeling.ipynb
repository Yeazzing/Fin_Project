```python
# IQR 방식을 이용한 이상치 제거 함수
def remove_outliers_iqr(df, column):
  Q1 = df[column].quantile(0.25)
  Q3 = df[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
  return df_filtered


# liabilities_ratio 컬럼부터 이상치 제거
for column in df.columns[4:]:  # 'liabilities_ratio' 컬럼부터 시작
  df1 = remove_outliers_iqr(df, column)

df1
```

## K-Means Clustering - 최적의 군집 수 찾기

```python
from sklearn.metrics import pairwise_distances

# Dunn Index 계산 함수
# Dunn Index
# 분모: 클러스터 내에서 가장 멀리 떨어진 두 점 사이의 거리 (Intra-cluster Distance) - 클러스터 내의 밀집도
# 분자: 클러스터 간의 가장 가까운 거리 (Inter-cluster Distance) - 클러스터 간의 분리도
# 값이 클수록 클러스터 간 분리가 잘 되고, 클러스터 내부는 잘 밀집되어 있다는 것을 의미

def dunn_index(data, labels):
    """
    Dunn Index를 계산하는 함수.
    :param data: 입력 데이터 (numpy array)
    :param labels: 각 데이터 포인트의 군집 레이블 (numpy array)
    :return: Dunn Index 값 (float)
    """
    unique_cluster_labels = np.unique(labels)
    num_clusters = len(unique_cluster_labels)

    if num_clusters == 1:
        return 0

    # 클러스터 간 최소 거리 계산
    inter_cluster_distances = []
    for i in range(num_clusters):
        for j in range(i + 1, num_clusters):
            cluster_i = data[labels == unique_cluster_labels[i]]
            cluster_j = data[labels == unique_cluster_labels[j]]
            inter_cluster_distances.append(np.min(pairwise_distances(cluster_i, cluster_j)))

    # 클러스터 내 최대 거리 계산
    intra_cluster_distances = []
    for i in range(num_clusters):
        cluster_i = data[labels == unique_cluster_labels[i]]
        intra_cluster_distances.append(np.max(pairwise_distances(cluster_i, cluster_i)))

    return np.min(inter_cluster_distances) / np.max(intra_cluster_distances)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

#군집수별 평가 지표 산출 함수
def cluster_and_evaluate(data, k_range):
    silhouette_scores = []
    for k in k_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        labels = kmeans.fit_predict(data)
        silhouette_avg = silhouette_score(data, labels)
        silhouette_scores.append(silhouette_avg)
    return silhouette_scores
```

```python
# Dunn Index와 Silhouette Score 출력
def print_scores(k_values, dunn_scores, silhouette_scores, criteria_name):
    print(f"\n{criteria_name}에 따른 군집 수에 따른 Dunn Index와 Silhouette Score")
    print("K\tDunn Index\tSilhouette Score")
    for k, dunn, silhouette in zip(k_values, dunn_scores, silhouette_scores):
        print(f"{k}\t{dunn:.5f}\t\t{silhouette:.5f}")
```

```python
# 각 기준에 따른 데이터 분할
financial = scaled_df[['liabilities_ratio', 'current_ratio', 'retention_rate']]
profitability = scaled_df[['gross_profit_rate', 'net_profit_rate', 'ROE']]
volatility = scaled_df[['VHF', 'updown', 'beta']]
growth = scaled_df[['sales_growth_rate', 'profit_change_rate']]
dividend = scaled_df[['cashdivide', 'dividend', 'divide_count']]
```

```python
#최적의 군집 수 탐색
import numpy as np
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

# 변수 초기화
k_values = range(2, 25)

# 시간 측정 시작
start_time = time.time()

# 각 기준에 대한 클러스터링 및 지표 계산
financial_silhouette_scores, financial_dunn_scores = cluster_and_evaluate(financial, k_values)
profitability_silhouette_scores, profitability_dunn_scores = cluster_and_evaluate(profitability, k_values)
volatility_silhouette_scores, volatility_dunn_scores = cluster_and_evaluate(volatility, k_values)
growth_silhouette_scores, growth_dunn_scores = cluster_and_evaluate(growth, k_values)
dividend_silhouette_scores, dividend_dunn_scores = cluster_and_evaluate(dividend, k_values)

# 시간 측정 종료
end_time = time.time()
total_time = end_time - start_time
```

```python
print_scores(k_values, financial_dunn_scores, financial_silhouette_scores, "재무 안정성")
print_scores(k_values, profitability_dunn_scores, profitability_silhouette_scores, "수익성")
print_scores(k_values, volatility_dunn_scores, volatility_silhouette_scores, "변동성")
print_scores(k_values, growth_dunn_scores, growth_silhouette_scores, "성장성")
print_scores(k_values, dividend_dunn_scores, dividend_silhouette_scores, "배당")
#실루엣 점수 기준 (dunn_scores-이상치에 예민함), 이상치 issue
#재무 안정성 -> 3
#수익성 -> 2
#변동성 -> 2
#성장성 -> 3
#배당 -> 4
```

## 최적의 군집 수로 K-Means 수행

```python
# 각 기준별로 군집 분석 수행 및 군집 중심 반환
def perform_clustering(data, column_names, n_clusters):
    clustering_model = KMeans(n_clusters=n_clusters, random_state=42)
    labels = clustering_model.fit_predict(data[column_names])
    return labels, clustering_model.cluster_centers_
```

```python
# financial, growth, quality에 대해  군집 분석 수행 및 중심 계산
f_labels, f_centers = perform_clustering(scaled_df, ['liabilities_ratio', 'current_ratio', 'retention_rate'], 3)
p_labels, p_centers = perform_clustering(scaled_df, ['gross_profit_rate', 'net_profit_rate', 'ROE'], 2)
v_labels, v_centers = perform_clustering(scaled_df, ['VHF', 'updown', 'beta'], 2)
g_labels, g_centers = perform_clustering(scaled_df, ['sales_growth_rate', 'profit_change_rate'], 3)
d_labels, d_centers = perform_clustering(scaled_df, ['cashdivide', 'dividend', 'divide_count'], 4)

# 결과 데이터프레임에 클러스터 라벨 추가
new_df = scaled_df.copy()
new_df['f_cluster'] = f_labels
new_df['p_cluster'] = p_labels
new_df['v_cluster'] = v_labels
new_df['g_cluster'] = g_labels
new_df['d_cluster'] = d_labels

new_df
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, pairwise_distances

# 평가 지표 저장할 리스트 초기화
criteria_list = ['f', 'p', 'v', 'g', 'd']
silhouette_scores = {'f': [], 'p': [], 'v': [], 'g': [], 'd': []}
dunn_scores = {'f': [], 'p': [], 'v': [], 'g': [], 'd': []}
data_dict = {'f': financial, 'p': profitability, 'v': volatility, 'g': growth, 'd': dividend}
```

```python
# Dunn Index 계산 함수
def dunn_index(data, labels):
    """
    Dunn Index를 계산하는 함수.
    :param data: 입력 데이터 (numpy array)
    :param labels: 각 데이터 포인트의 군집 레이블 (numpy array)
    :return: Dunn Index 값 (float)
    """
    unique_cluster_labels = np.unique(labels)
    num_clusters = len(unique_cluster_labels)

    if num_clusters == 1:
        return 0

    # 클러스터 간 최소 거리 계산
    inter_cluster_distances = []
    for i in range(num_clusters):
        for j in range(i + 1, num_clusters):
            cluster_i = data[labels == unique_cluster_labels[i]]
            cluster_j = data[labels == unique_cluster_labels[j]]
            inter_cluster_distances.append(np.min(pairwise_distances(cluster_i, cluster_j)))

    # 클러스터 내 최대 거리 계산
    intra_cluster_distances = []
    for i in range(num_clusters):
        cluster_i = data[labels == unique_cluster_labels[i]]
        intra_cluster_distances.append(np.max(pairwise_distances(cluster_i, cluster_i)))

    return np.min(inter_cluster_distances) / np.max(intra_cluster_distances)
```

```python
#최적의 군집 수 탐색
k_values = range(2, 25)

# 지표 저장 딕셔너리 초기화
silhouette_scores_single = {criteria: [] for criteria in criteria_list}
dunn_scores_single = {criteria: [] for criteria in criteria_list}
silhouette_scores_complete = {criteria: [] for criteria in criteria_list}
dunn_scores_complete = {criteria: [] for criteria in criteria_list}

# 각 기준에 대해 클러스터링 및 지표 계산 (Single Linkage)
for criteria in criteria_list:
    data = data_dict[criteria].values  # NumPy 배열로 변환
    for k in k_values:
        clusterer_single = AgglomerativeClustering(n_clusters=k, linkage='single')
        cluster_labels_single = clusterer_single.fit_predict(data)
        silhouette_avg_single = silhouette_score(data, cluster_labels_single)
        dunn_avg_single = dunn_index(data, cluster_labels_single)
        silhouette_scores_single[criteria].append(silhouette_avg_single)
        dunn_scores_single[criteria].append(dunn_avg_single)

# 각 기준에 대해 클러스터링 및 지표 계산 (Complete Linkage)
for criteria in criteria_list:
    data = data_dict[criteria].values  # NumPy 배열로 변환
    for k in k_values:
        clusterer_complete = AgglomerativeClustering(n_clusters=k, linkage='complete')
        cluster_labels_complete = clusterer_complete.fit_predict(data)
        silhouette_avg_complete = silhouette_score(data, cluster_labels_complete)
        dunn_avg_complete = dunn_index(data, cluster_labels_complete)
        silhouette_scores_complete[criteria].append(silhouette_avg_complete)
        dunn_scores_complete[criteria].append(dunn_avg_complete)
```

```python
from sklearn.cluster import DBSCAN

# 최적의 하이퍼파라미터를 찾기 위한 함수
def find_best_dbscan_params(data, column_names):
    best_eps = None
    best_minPts = None
    best_num_clusters = 0
    best_silhouette = -1
    best_combinations = []

    eps_values = np.arange(0.5, 3.0, 0.2)
    minPts_values = range(2, 25)

    for eps in eps_values:
        for minPts in minPts_values:
            dbscan = DBSCAN(eps=eps, min_samples=minPts)
            labels = dbscan.fit_predict(data[column_names])

            num_clusters = len(set(labels)) - (1 if -1 in labels else 0) #이상치: -1

            if num_clusters > 1:  # 유효한 군집이 존재할 때만 평가
                silhouette_avg = silhouette_score(data[column_names], labels)
                if silhouette_avg > best_silhouette:
                    best_eps = eps
                    best_minPts = minPts
                    best_num_clusters = num_clusters
                    best_silhouette = silhouette_avg
                    best_combinations = [(eps, minPts, silhouette_avg)]
                elif silhouette_avg == best_silhouette:
                    best_combinations.append((eps, minPts, silhouette_avg))

    return best_eps, best_minPts, best_num_clusters, best_silhouette, best_combinations
```

### 재 클러스터링

1. 각 클러스터의 조합을 벡터로 표현
2. 그것들 간의 유사도를 계산하여 임계값에 따라 클러스터를 묶는 방식 (코사인 유사도 사용하여 유사한 군집끼리 묶어줌)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
from sklearn.metrics import pairwise_distances
import numpy as np

# LabelEncoder 객체 생성
le = LabelEncoder()

# 각 컬럼에 대해 레이블 인코딩 수행
for column in ['financial_stability', 'profitability', 'volatility', 'growth', 'dividend_stability']:
  df[column] = le.fit_transform(df[column])

# 벡터화된 데이터 확인
vectorized_data = df[['financial_stability', 'profitability', 'volatility', 'growth', 'dividend_stability']].values

#유사도 계산
#similarity_matrix = cosine_similarity(vectorized_data) #코사인
# distance_matrix = euclidean_distances(vectorized_data) #유클리드
# similarity_matrix = 1 / (1 + distance_matrix)
distance_matrix = pairwise_distances(vectorized_data, metric='hamming')
similarity_matrix = 1 - distance_matrix

# 임계값 설정 (예: 0.9 이상이면 유사하다고 판단)
# 이 값을 낮게 설정하면 group 개수가 적어지고, 높게 설정하면 많아짐
# 0.8로 하면 5개, 0.9로 하면 14개 나옴
threshold = 0.6
```

```python
# 유사한 클러스터끼리 그룹핑
n_clusters = vectorized_data.shape[0]
groups = [-1] * n_clusters  # -1로 초기화된 그룹 리스트
group_id = 0

for i in range(n_clusters):
    if groups[i] == -1:  # 아직 그룹에 속하지 않은 클러스터
        groups[i] = group_id  # 새로운 그룹 ID 할당
        for j in range(i + 1, n_clusters):
            if similarity_matrix[i][j] >= threshold:  # 유사도 임계값 이상인 경우
                groups[j] = group_id
        group_id += 1

# 결과를 데이터프레임에 추가
df['group'] = groups

# 유사한 클러스터끼리 묶인 결과 출력
df
```

* 작은 그룹 합치기 & 너무 큰 그룹 나누기

```python
from collections import Counter

group_sizes = Counter(groups)

# Identify groups to merge (fewer than 40 members)
small_groups = [group for group, size in group_sizes.items() if size < 40]

# Function to find the closest group to merge based on similarity matrix
def find_closest_group(small_group, groups, similarity_matrix):
    closest_group = None
    max_similarity = -1

    # Find the most similar group that is not the small group itself
    for i in range(len(groups)):
        if groups[i] == small_group:
            for j in range(len(groups)):
                if groups[j] != small_group and similarity_matrix[i][j] > max_similarity:
                    closest_group = groups[j]
                    max_similarity = similarity_matrix[i][j]

    return closest_group

# Continuously merge small groups until none remain
while len(small_groups) > 0:
    for small_group in small_groups:
        closest_group = find_closest_group(small_group, groups, similarity_matrix)

        # If no similar group is found, skip (shouldn't happen)
        if closest_group is None:
            continue

        # Merge all members of the small group into the closest group
        for i in range(len(groups)):
            if groups[i] == small_group:
                groups[i] = closest_group

    # Recalculate group sizes and find new small groups
    group_sizes = Counter(groups)
    small_groups = [group for group, size in group_sizes.items() if size < 40]

# Split large groups if necessary
new_group_id = max(groups) + 1
large_groups = [group for group, size in group_sizes.items() if size > 100]
for large_group in large_groups:
    indices = [i for i, group in enumerate(groups) if group == large_group]

    # Split the large group into subgroups of around 100 items each
    for start in range(0, len(indices), 100):
        sub_indices = indices[start:start + 100]
        new_id = large_group if start == 0 else new_group_id
        for idx in sub_indices:
            groups[idx] = new_id

        if start != 0:
            new_group_id += 1

# Update the DataFrame with the new merged and split groups
df['group'] = groups

# Output the adjusted grouped data
print(df['group'].value_counts())
```
